## Compare models
Test the baseline models. Choose based on the parameters below:

Accuracy- The fraction of predictions our model got right

AUC-  Based on the graph

Recall - What proportion of actual positives was identified correctly. i.e. fraction of relevant information that are retrieved.    TP/(TP+FN)

Precision -  What proportion of positive identifications was actually correct. i.e. the fraction of information retrieved that are relevant.   TP/(TP+FP)

F1 - harmonic mean of precision and recall taking both metrics into account  2 (Precision*Recall)/(Precision+Recall)

Kappa -
<0 No agreement
0 — .20 Slight
.21 — .40 Fair
.41 — .60 Moderate
.61 — .80 Substantial
.81–1.0 Perfect

<br/>
